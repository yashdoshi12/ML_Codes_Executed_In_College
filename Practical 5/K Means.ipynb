```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('diabetes.csv')

# Exclude the 'Outcome' column
X = data.drop('Outcome', axis=1)
```

```python
# Determine the optimal number of clusters (K)
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Plot the WCSS values
plt.plot(range(1, 11), wcss)
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```

```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt

# Read the diabetes CSV file
df = pd.read_csv("diabetes.csv")

plt.scatter(df.Age, df.Glucose)
plt.xlabel('Age')
plt.ylabel('Glucose')
plt.show()


km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age', 'Glucose']])


df['cluster'] = y_predicted


cluster_centers = km.cluster_centers_


df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]


plt.scatter(df1.Age, df1.Glucose, color='green', label='Cluster 0')
plt.scatter(df2.Age, df2.Glucose, color='red', label='Cluster 1')
plt.scatter(df3.Age, df3.Glucose, color='black', label='Cluster 2')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], color='purple', marker='*', label='Centroid')
plt.xlabel('Age')
plt.ylabel('Glucose')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt

# Read the diabetes CSV file
df = pd.read_csv("diabetes.csv")

# Select the features for clustering
X = df[['Age', 'Glucose']]

# Perform K-means clustering for different values of k
inertia = []
k_values = range(1, 11)  # Trying k values from 1 to 10

for k in k_values:
    km = KMeans(n_clusters=k)
    km.fit(X)
    inertia.append(km.inertia_)

# Plotting the elbow curve
plt.plot(k_values, inertia, 'bx-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

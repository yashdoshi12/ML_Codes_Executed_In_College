```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("diabetes.csv")

print(df.head())
```

```python
df['Outcome'].plot(kind='hist')
plt.title('Distribution of Outcome')
plt.xlabel('Outcome Value')
plt.show()

df.plot(x='Glucose', y='BMI', kind='scatter')
plt.title('Glucose vs BMI')
plt.xlabel('Glucose')
plt.ylabel('BMI')
plt.show()
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split

X = df.drop('Outcome', axis=1)
y = df['Outcome']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

clf = tree.DecisionTreeClassifier()

clf.fit(X_train, y_train)

accuracy = clf.score(X_test, y_test)
print("Accuracy:", accuracy)
```

```python
import pandas as pd
from sklearn import tree
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.model_selection import train_test_split


diabetes = pd.read_csv("diabetes.csv")


X = diabetes.drop('Outcome', axis=1)
y = diabetes['Outcome']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


k_best = SelectKBest(score_func=chi2, k=5)
X_train = k_best.fit_transform(X_train, y_train)
X_test = k_best.transform(X_test)

clf = tree.DecisionTreeClassifier()

clf.fit(X_train, y_train)


accuracy = clf.score(X_test, y_test)
print("Accuracy with feature selection using chi-squared test:", accuracy)
```

```python
import pandas as pd
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


diabetes = pd.read_csv("diabetes.csv")


X = diabetes.drop('Outcome', axis=1)
y = diabetes['Outcome']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)


clf = tree.DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=0)

clf.fit(X_train, y_train)

accuracy = clf.score(X_test, y_test)
print("Accuracy:", accuracy)

tree_text = tree.plot_tree(clf.fit(X_train, y_train))
print(tree_text)
```
